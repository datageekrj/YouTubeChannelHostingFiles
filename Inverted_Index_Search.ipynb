```python
import nltk
from nltk.corpus import shakespeare, stopwords
```

```python
nltk.download("shakespeare")
nltk.download("stopwords")
```

```python
shakespeare.fileids()
```

```python
# {
#     word1 : set() of document_ids which has the word1,
#     word2 : set() of document_ids which has the word1,
#     word3 : set() of document_ids which has the word1
#     .....
# }
```

```python
stop_words_english = stopwords.words("english")
```

```python
"shgs".isalnum()
```

```python
inverted_index = {}
```

```python
for doc_id, doc_name in enumerate(shakespeare.fileids()):
  raw_words = shakespeare.words(doc_name)
  preprocessed_words = [word.lower() for word in raw_words if word.isalnum() and word not in stop_words_english]
  for word in preprocessed_words:
    if word in inverted_index:
      inverted_index[word].add(doc_id)
    else:
      inverted_index[word] = {doc_id}
```

```python
query = "tragedy of cleopatra"
```

```python
def search_index(index, query):
  words = query.split(" ")
  preprocessed_words = [word for word in words if word.isalnum() and word not in stop_words_english]

  list_of_doc_ids = []
  for word in preprocessed_words:
    list_of_doc_ids.append(index.get(word, set()))

  return set.intersection(*list_of_doc_ids)
```

```python
search_index(inverted_index, query)
```
